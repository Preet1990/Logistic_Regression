```python
from sklearn.datasets import load_breast_cancer
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
```

```python

#Load the data set

cancer = load_breast_cancer()
df_cancer = pandas.DataFrame(cancer.data,columns=cancer.feature_names)
df_cancer['label'] = pandas.Series(cancer.target)
df_cancer.head()
```

```python
df_cancer.describe()
```

```python
df_cancer.info()
```

```python
# X = feature values, all the columns except the label column

X = df_cancer.iloc[:, :-1]
display(X)
```

```python
# y = target values, only the label column

Y = df_cancer.iloc[:, -1]
display(Y)
```

```python
import plotly.express as px
```

```python
# plot histogram plot

plt.xlabel('Feature')
plt.ylabel('Label')

df_cancer['label'].hist()


plt.show()
```

```python

import plotly.express as px
fig = px.scatter(df_cancer, x='mean radius', y='label',color='label')
fig.show()
```

```python
#Split the data into 80% training and 20% testing

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
#Train the model
model = LogisticRegression(verbose=1)
model.fit(X_train, Y_train)
```

```python
#Test the model

predictions = model.predict(X_test)
print(predictions)   # printing predictions

print()# Printing new line

#Check precision, recall, f1-score
print( classification_report(Y_test, predictions) )
```

```python
# Confusion matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, predictions)
```
